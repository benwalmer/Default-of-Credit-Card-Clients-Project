Notice X6-11, X12-17, X18-23 are all measuring of the same thing across different months. We decided to make variables that average across each feature for simplicity.

```python
X["X6-X11"] = X[["X6", "X7", "X8", "X9", "X10", "X11"]].mean(axis = 1)
X["X12-X17"] = X[["X12", "X13", "X14", "X15", "X16", "X17"]].mean(axis = 1)
X["X18-23"] = X[["X18", "X19", "X20", "X21", "X22", "X23"]].mean(axis = 1)
```

It would be wise to normalize the features so they are all on the same scale

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_features = X.copy()
X_features = X_features.drop(columns = ["X6", "X7", "X8", "X9", "X10", "X11", "X12", "X13", "X14", "X15", "X16", "X17", "X18", "X19", "X20", "X21", "X22", "X23"])
X_scaled = pd.DataFrame(scaler.fit_transform(X_features), columns = X_features.columns)
X_scaled.describe()
```

```python
corr = X_scaled.corr().abs()
s = corr.unstack()
so = s.sort_values(ascending = False, kind="quicksort")
so[(so < 1) & (so >= .2)]
```

It seems that many of the features are correlated with one another, especially features X12-X17. This makes sense because these features are the amounts of different bill statements in different months in 2005, so the bill of one month will be pretty similar to the bill from the previous month. For now we will not remove any variables, as there is no perfect correlation of 1 between any two features. We can remove features if there become problems with multi-collinearity at a later step.

```python
corrs = pd.concat([X_scaled, y], axis = 1)
corr2 = corrs.corr().abs()
corr2['Y'].sort_values(ascending = False)
```

We see that X6-X11, features that measure the history of past payment, have the highest correlation with y, whether a person defaults on their payment or not, which makes sense. Additionally X1, the amount of credit, and X18-23, the amount of previous payments, also have the next highest correlation with y.

We decided to keep all features for our models as there were only 8 features total.

```python
corrfinal = X_scaled.corr()
import seaborn as sns
sns.heatmap(data = corrfinal, annot = True, annot_kws = {"size": 6})
```

No huge problems with multicollinearity

Model 1: Plain Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.25, random_state=21)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

dt_pipe = Pipeline([('dt', DecisionTreeClassifier())])
params = [{'dt__max_depth': [3, 4, 5, 6],
         'dt__min_samples_leaf': [2, 3, 4, 5, 6]}]

gs_dt = GridSearchCV(dt_pipe,
                      param_grid=params,
                      scoring='accuracy',
                      cv=5)
gs_dt.fit(X_train, y_train)
print(gs_dt.best_params_)
# Find best model score
print(gs_dt.score(X_test, y_test))
dt_best = gs_dt.best_estimator_.named_steps['dt']
```

```python
# Export tree to Graphviz format
dot_data = export_graphviz(
    dt_best,
    out_file=None,
    feature_names=X_train.columns,
    class_names=[str(c) for c in dt_best.classes_],  # convert to strings
    filled=True, rounded=True,
    special_characters=True
)

# Render with graphviz
graph = graphviz.Source(dot_data)
graph
```

```python
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay,
    roc_curve, auc, RocCurveDisplay,
    precision_recall_curve, PrecisionRecallDisplay,
    accuracy_score, precision_score, recall_score, f1_score
)
import matplotlib.pyplot as plt
import numpy as np

# ---- Confusion Matrix ----
y_pred = gs_dt.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot()
plt.title("Confusion Matrix")
plt.show()

# ---- ROC Curve ----
y_prob = gs_dt.predict_proba(X_test)[:, 1]  # probability of positive class

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random Guess')  # baseline
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# ---- Precision-Recall Curve ----
prec, rec, _ = precision_recall_curve(y_test, y_prob)
no_skill = sum(y_test.values) / len(y_test)  # baseline = positive class ratio

plt.figure(figsize=(6, 5))
plt.plot(rec, prec, color='green', label='Precision-Recall Curve')
plt.hlines(no_skill, 0, 1, color='gray', linestyle='--', label='Random Guess')  # baseline
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# ---- Metrics ----
acc = accuracy_score(y_test, y_pred)
prec_val = precision_score(y_test, y_pred)
rec_val = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Model Evaluation Metrics")
print("--------------------------")
print(f"Accuracy:  {acc:.3f}")
print(f"Precision: {prec_val:.3f}")
print(f"Recall:    {rec_val:.3f}")
print(f"F1 Score:  {f1:.3f}")
print(f"AUC:       {roc_auc:.3f}")
```

Model 4: XG Boost

```python
# Imports
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc,
    precision_recall_curve, average_precision_score
)
import matplotlib.pyplot as plt
```

```python
# --- Base model setup ---
xgb = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=21
)

# --- Parameter grid for tuning ---
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# --- GridSearchCV ---
grid_search_xgb = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=2
)

# --- Fit the model ---
grid_search_xgb.fit(X_train, y_train)

# --- Results ---
print("âœ… Best Parameters for XGBoost:")
print(grid_search_xgb.best_params_)
print(f"Best Cross-Validation F1 Score: {grid_search_xgb.best_score_:.4f}")

# --- Best model ---
best_xgb = grid_search_xgb.best_estimator_

# --- Predictions ---
y_pred_xgb = best_xgb.predict(X_test)
y_proba_xgb = best_xgb.predict_proba(X_test)[:, 1]

# --- Evaluation ---
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb))

# --- ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_proba_xgb)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.3f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - XGBoost Classifier")
plt.legend()
plt.show()

# --- Precision-Recall Curve ---
prec, rec, _ = precision_recall_curve(y_test, y_proba_xgb)
ap = average_precision_score(y_test, y_proba_xgb)

plt.figure(figsize=(6, 5))
plt.plot(rec, prec, label=f"AP = {ap:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve - XGBoost Classifier")
plt.legend()
plt.show()

# --- Feature Importance ---
plt.figure(figsize=(8, 6))
plot_importance(best_xgb, max_num_features=10, importance_type="gain")
plt.title("Top 10 Feature Importances (XGBoost - Gain)")
plt.show()
```
